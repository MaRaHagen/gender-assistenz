```python
from word_usage_analyzer.igw import is_gendered_word, gendered_form_to_feminine_form
from spacy.tokens.token import Token

Token.set_extension("was_gendered", default=False, force=True)
Token.set_extension("original_word", default="", force=True)

def femininized_spacy(text):
    doc = spacify(text)

    replaced = {

    }

    newstr = ""
    for sent in doc.sents:
        for word in sent:
            if is_gendered_word(word.text):
               newstr = newstr + " " + gendered_form_to_feminine_form(word.text)
               replaced[gendered_form_to_feminine_form(word.text)] = word.text
            else:
               newstr = newstr + (" " if not word.is_punct else "") + word.text

    doc2 = spacify(newstr)

    for sent in doc2.sents:
        for word in sent:
            word._.was_gendered = replaced.get(word.text) is not None
            word._.original_word = replaced.get(word.text)

    return doc2

doc = femininized_spacy(example_text)

print(doc)
for sent in doc.sents:
    for word in sent:
        print(f"{word._.was_gendered}: {word.text} {word.lemma_} {word.morph} ")

```
